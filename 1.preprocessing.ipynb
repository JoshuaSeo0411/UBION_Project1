## 흥행 영화 데이터만 추출
* 흥행 영화 : 매출액 > 평균제작비

```python
# index 0번 삭제 및 순번 컬럼 삭제
df.drop('순번', axis=1, inplace=True, errors='ignore')
```

```python
# 연도만 추출하여 새로운 파생변수 개봉연도 생성
df['개봉일'] = pd.to_datetime(df['개봉일'])
df['개봉연도'] = df['개봉일'].dt.year
```

```python
# 새로운 파생변수 평균 제작비 컬럼 생성
def change(x):
    if x == 2009:
        result = 23.1
    elif x == 2010:
        result = 21.6
    elif x == 2011:
        result = 22.7
    elif x == 2012:
        result = 20.3
    elif x == 2013:
        result = 21.4
    elif x == 2014:
        result = 20.1
    elif x == 2015:
        result = 19.9
    elif x == 2016:
        result = 24.0
    elif x == 2017:
        result = 26.3
    elif x == 2018:
        result = 26.8
    elif x == 2019:
        result = 29.0
    else:
        result = '!'
    return result

df['평균 제작비'] = df['개봉연도'].apply(change)
```

```python
df.head()
```

```python
# 매출액 억단위로 변경
df['수정전국매출액']  =  df['전국매출액'] / 100000000
```

```python
# 평균 제작비보다 매출액이 더 높은 데이터만 추출
df = df[df['수정전국매출액'] >= df['평균 제작비']]
```

```python
df.reset_index(drop='index', inplace=True)
```

```python
df.drop(['수정전국매출액'], axis=1, inplace=True)
```

# 주식지수 삽입
* 개봉일 시가, 판매일 수정종가

```python
## 데이터 로드

# 업종지수
movie_index = pd.read_csv('../datasets/movie_index.csv', index_col=0)
# 국고채10년 금리 데이터
market_rate = pd.read_csv('../datasets/market_rate.csv')
```

```python
## 일자 시계열데이터로 변경, 수정주가지수, 12일 뒤 주가지수 컬럼 생성
movie_index['일자'] = pd.to_datetime(movie_index['일자'])

movie_index['수정주가지수12'] = movie_index['수정주가지수'].shift(11)
movie_index.columns = ['개봉일', '수정주가지수', '수정주가지수12']

movie_index.head()
```

```python
## 개봉일 기준 주가지수, 수정주가지수 merge, 결측치(주말, 공휴일 등) interpolate 처리

new_df['개봉일'] = pd.to_datetime(new_df['개봉일'])
new_df = new_df.reset_index(drop='index')
new_df = pd.merge(new_df, movie_index, how='left', on='개봉일')
new_df['수정주가지수'] = new_df['수정주가지수'].interpolate()
new_df['수정주가지수12'] = new_df['수정주가지수12'].interpolate()
```

```python
## 국고채금리 데이터 수정
market_rate = market_rate.drop(['통계표', '단위', '변환'], axis=1)
market = market_rate.T
market = market.reset_index()
market.columns = ["개봉연도", "KORIBOR(12개월)","CD(91일)","CP(91일)","국고채(10년)"]
market = market.iloc[1:]
market['개봉연도'] = market['개봉연도'].astype(int)
```

```python
market['KORIBOR(12개월)'] = market['KORIBOR(12개월)'].astype(float)
market['CD(91일)'] = market['CD(91일)'].astype(float)
market['CP(91일)'] = market['CP(91일)'].astype(float)
market['국고채(10년)'] = market['국고채(10년)'].astype(float)
```

```python
market_bond = market[['개봉연도', '국고채(10년)']]
market_bond
```

```python
## 국고채 금리 원본 데이터에 merge로 결합
new_df = pd.merge(new_df, market_bond, how='left', on='개봉연도')
```

```python
## 수익률 컬럼 생성
# 12일 수익률 컬럼 생성
new_df['12일수익률'] = (new_df['수정주가지수12'] - new_df['수정주가지수']) / new_df['수정주가지수']

# 일일 수익률 컬럼 생성
new_df['일일수익률'] = (new_df['12일수익률'] + 1) ** (1/12) - 1

# 연간 수익률 컬럼 생성
new_df['연간수익률'] = ((1 + new_df['일일수익률']) ** 252 - 1) * 100
```

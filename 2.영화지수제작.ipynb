```python
# 시가총액주가지수 제작
merdf['시가총액주가지수'] = merdf['시가총액'] / 688191757200.0 * 100
```

```python
# 연도별 기업개수 컬럼 생성
merdf['기업개수'] = 14 - mer_int.isna().sum(1)
```

```python
# 수정주가지수 제작(기업개수 * 4는 기준값을 100으로 맞추기 위함)
merdf['수정주가지수'] = merdf['시가총액주가지수'] / merdf['기업개수'] * 4
```

## 볼린저 밴드 제작(표준편차 1.5)
---

```python
movie_index = pd.read_csv('../datasets/영화업종지수.csv')
```

```python
## 볼린저 밴드 중심선 제작(20일 이동평균선)
movie_index["center"] = movie_index['수정주가지수'].rolling(20).mean()
```

```python
## 결측치 선형보간법으로 채우기
movie_index["center"] = movie_index['center'].interpolate()
```

```python
## 볼린저 밴드 ub 제작
movie_index["ub"] = \
    movie_index["center"] + ( 1.5 * movie_index["수정주가지수"].rolling(20).std())
```

```python
## 볼린저 밴드 lb 제작
movie_index["lb"] = \
    movie_index['center'] - ( 1.5 * movie_index['수정주가지수'].rolling(20).std() )
```

```python
movie_index
```

```python
## 볼린저 밴드(1.5표준편차) 데이터 저장
movie_index.to_csv('../datasets/bolinger(1.5).csv')
```

```python
movie_index['일자'] = pd.to_datetime(movie_index['일자'])
```

```python
## 볼린저 밴드 시각화
plt.figure(figsize=(20, 10))
plt.plot(movie_index.index, movie_index['수정주가지수'], label='수정주가지수')
plt.plot(movie_index.index, movie_index['ub'], linestyle='dashed', label='Upper band')
plt.plot(movie_index.index, movie_index['center'], linestyle='dashed', label='Moving Average 20')
plt.plot(movie_index.index, movie_index['lb'], linestyle='dashed', label='Lower band')
plt.title('영화업종지수의 볼린저 밴드(20일, 1.5 표준편차)')
plt.legend(loc='best');
```
